```python
import numpy as np
import matplotlib.pyplot as plt

# Training data
x = np.array([1, 2, 3])
y = np.array([2, 4, 6])

# Parameter ranges
w_vals = np.linspace(-1, 5, 1000)
b_vals = np.linspace(-1, 5, 1000)

# Create grid
W, B = np.meshgrid(w_vals, b_vals)

# Cost function (Mean Squared Error)
J = np.zeros(W.shape)

for i in range(len(w_vals)):
    for j in range(len(b_vals)):
        y_pred = W[j, i] * x + B[j, i]
        J[j, i] = np.mean((y_pred - y) ** 2)

# Draw contour plot
plt.contour(W, B, J, levels=20)
plt.xlabel("w (slope)")
plt.ylabel("b (intercept)")
plt.title("Contour Plot of Cost Function")
plt.colorbar()
plt.show()
```
