## NumPy Indexing and Selection

```python
arr = np.arange(0,11) # create 1-dimentional array from 0 to 10
```

```python
arr
```

```python
arr[1]
```

```python
arr[1:5]
```

```python
arr[0:6]
```

```python
arr[5:]
```

```python
arr[0:5] = 100
```

```python
arr
```

```python
slice_of_arr = arr[0:6]
```

```python
slice_of_arr
```

```python
arr_copy = arr.copy()
```

```python
arr_copy
```

```python
arr_copy[:] = 100
```

```python
arr_copy # initial arr remains unchanged
```

```python
#### 2D array

arr_2d = np.array([[5,10,15], [20,25,30], [35,40,45]])
```

```python
print(arr_2d, arr_2d.shape)
```

```python
arr_2d[0][0]
```

```python
arr_2d[0]
```

```python
print(arr_2d[1][1], arr_2d[1,1]) # both work
```

```python
arr_2d[:2,1:]
```

```python
arr = np.arange(1,11)
```

```python
arr
```

```python
bool_arr = arr > 5 # boolean array: True for elements > 5, False otherwise
```

```python
bool_arr
```

```python
arr[bool_arr] # filtering using boolean array
```

```python
arr[arr > 5] # conditional selection   
```

```python
arr[arr < 3]
```

```python
arr_2d = np.arange(50).reshape(5,10)
```

## NumPy Opeartions

```python
arr = np.arange(1,11)
```

```python
arr
```

```python
arr + arr # addition    
```

```python
arr * arr # multiplication
```

```python
arr - arr # subtraction
```

```python
arr + 100
```

```python
arr - 1
```

```python
arr
```

```python
arr ** 2
```

```python
np.sqrt(arr)
```

```python
np.amax(arr)
```

```python
# An exponential function represents the relationship between an input and ouput,
# where we use repreated multiplication on an initial value to get the output for any 
# given input. Exponential can grow or decay very quickly.

# y = e^x --> `e`` is the natural base, `x` is the element in the array
print(arr, np.exp(arr))
```
